## fixing the seed value for the random selection guarantees the same results in repeated runs

```r
set.seed(1) # 
```

```r
s = sample(nrow(auto),floor(nrow(auto)*.6))
```

```r
s = sort(s)
s
```

```r
train <- auto[s,] # 무작위 행 추출
eval <- auto[-s,] # train 데이터의 행 빼고 추출
```

```r
eval[1:3,] #1부터 3행까지만 추출
```

```r
train[1:3,]
```

## regression on training set

```r
m2=lm(Price~.,data=train)
summary(m1)
summary(m2)
```

```r
eval$Prediction = predict(m2, eval) # linear regression 예측해보기
```

```r
head(eval)
```

```r
d = mean(eval$Price - eval$Prediction) #Mean Error
d
```

```r
sqrt(mean((eval$Price - eval$Prediction)^2)) #RMSE(root mean square error)
```

```r
a = mean(abs(eval$Price - eval$Prediction)/eval$Price) #MAPE(Mean absolute percent error)
a
```
